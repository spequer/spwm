```
base_period = 50

# main freq 20k - 25k
main_min = 20000
main_max = 25000

# target sin 10hz - 1khz
sin_min =  10
sin_max =  1000

# points
points = 512

  
from math import log, exp, sqrt

debug = 0

lsmax = log(sin_max)
lsmin = log(sin_min)
bp = base_period*10**-9
op_min = 1.0/main_max/bp
op_max = 1.0/main_min/bp

#print('op min/max', op_min, op_max)
#csvf.writerow(['Target Freq', 'Opening', 'Step', 'Real Freq', 'Error%'])
if not debug:
   print(','.join(['Target Freq', 'Opening', 'Step', 'Real Freq', 'Error%']))

for n, i in enumerate(range(0, points)):
    lsx = (lsmax-lsmin) / (points-1) * i + lsmin
    tarf = exp(lsx)
    if debug:
      print('n:', n, tarf)
    sx = exp(lsx)
    psx = 1.0/sx
    psxm = psx / bp
    psxq = sqrt(psxm)
    if debug:
      print("psx:", psx, psxq)
    if psxq < op_min:
        psxq = op_min
    if psxq > op_max:
        psxq = op_max

    op = int(psxq)
    if debug:
      print('opening:', op)
    rmain_freq = 1.0/(bp*op)
    if debug:
      print('rmain_freq:' , rmain_freq)
    xstep = psx / (1.0/rmain_freq)
    xstepi = int(xstep)
    if debug:
      print('xstep:', xstepi, xstep)
    real_op = int((xstep / xstepi) * op)
    if debug:
      print('real_op', real_op)
      print('-'*80)
    real_freq = 1.0/(real_op*xstepi*bp)
    if debug:
      print('rfreq', tarf, real_freq)
      print('='*80)
    err = ((real_freq/tarf)-1) * 100
    #if n >= 500:
    #    break
    #csvf.writerow([f'{tarf:.2f}', f'{real_op}', f'{xstepi}', f'{real_freq:.2f}', f'{err:.2f}%'])
    if not debug:
      print(','.join([f'{tarf:.2f}', f'{real_op}', f'{xstepi}', f'{real_freq:.2f}', f'{err:.2f}%']))
    


```
